```python
# Get All pages Phones
names = []
prices = []
for i in range(36):
    url = f'https://www.jumia.com.eg/mobile-phones/?page={i+1}#catalog-listing'
    respons = requests.get(url)
    suop = BeautifulSoup(respons.text , 'html.parser')
    
    info = suop.find_all('div' , attrs = {'class':'info'})
    
    for i in info:
        name = i.find('h3').get_text()
        names.append(name)
        price = i.find('div' , attrs = {'class':'prc'}).get_text()
        prices.append(price)
```
